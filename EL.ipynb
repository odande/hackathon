```python
import json
import pandas as pd
import requests

link="https://results.elections.europa.eu/data-sheets/json/2019-2024/election-results/el.json"
r=requests.get(link)
EL_2019=json.loads(r.text)
```

```python
df_EL_2019=pd.json_normalize(EL_2019['groupSummary']['groupDistribution'])
```

```python
df_EL_2019
```

```python
df_EL_2019['periode'] = '2019-2024'
```

```python
link1="https://results.elections.europa.eu/data-sheets/json/2024-2029/election-results/el.json"
r=requests.get(link1)
EL_2024=json.loads(r.text)
```

```python
df_EL_2024 = pd.json_normalize(EL_2024['groupSummary']['groupDistribution'])
```

```python
df_EL_2024['periode'] = '2024-2029'
```

```python
df_EL_2024
```

```python
link3 = 'https://results.elections.europa.eu/data-sheets/json/2024-2029/election-results/groups.json'
r3 = requests.get(link3)

groups_2024 = json.loads(r3.text)
```

```python
df_groups_2024 = pd.json_normalize(groups_2024['groups'], record_path='translations', meta='groupId')
```

```python
df_groups_2024
```

```python
el_groups_2024 = df_groups_2024.loc[df_groups_2024['languageId'] == 'EL', ['groupLongName','groupId']]
```

```python
df_EL_2024 = pd.merge(df_EL_2024, el_groups_2024, how='left', left_on='id', right_on='groupId')
```

```python
df_EL_2024
```

```python
link4="https://results.elections.europa.eu/data-sheets/json/2019-2024/outgoing-parliament/groups.json"
r4 = requests.get(link4)
groups_2019 = json.loads(r4.text)
```

```python
df_groups_2019 = pd.json_normalize(groups_2019['groups'], record_path='translations', meta='groupId')
```

```python
el_groups_2019 = df_groups_2019.loc[df_groups_2019['languageId'] == 'EL', ['groupLongName','groupId']]
```

```python
df_EL_2019 = pd.merge(df_EL_2019, el_groups_2019, how='left', left_on='id', right_on='groupId')
```

```python
df_EL_2019
```

```python
df_EL_2024 = df_EL_2024.drop(columns='id')
df_EL_2019 = df_EL_2019.drop(columns='id')
```

```python
df_EL_2024[['groupLongName', 'groupId']].value_counts()
```

```python
df_EL_2019[['groupLongName', 'groupId']].value_counts()
```

```python
df_EL_2024 = df_EL_2024.drop(columns='periode')
df_EL_2019 = df_EL_2019.drop(columns='periode')
```

```python
df_EL = pd.merge(df_EL_2019, df_EL_2024, how='left', left_on='groupId', right_on='groupId')
```

```python
df_EL
```
