```python
import requests
import json

def geocode(adresse):
    url = f"https://nominatim.openstreetmap.org/search?addressdetails=1&q={adresse}&format=jsonv2"
    response = requests.get(url)
    geo = response.json()
    print(geo)
    if geo:
        latitude = geo[0]['lat']
        longitude = geo[0]['lon']
        return [latitude, longitude]
    else:
        return "Adresse non trouvée"

```
